Q 1: What is the total profit generated?

```python
total_profit = df["Profit"].sum()
print(f"Total profit generated: ${total_profit:.2f}")
```

Q 2: Which product category contributed the most to total sales?



```python
top_category = df.groupby("Category")["Sales"].sum().idxmax()
category_sales = df.groupby("Category")["Sales"].sum().max()
print(f"Category with highest sales: {top_category} (${category_sales:.2f})")
```

Q 3: What is the average discount offered across all orders?

```python
average_discount = df["Discount"].mean()
print(f"Average discount offered: {average_discount:.2%}")
```

Q 4: Which ship mode is the most commonly used?

```python
# Assuming your dataframe is named df
ship_mode_counts = df['Ship Mode'].value_counts()

# Get the most commonly used ship mode
most_common_ship_mode = ship_mode_counts.idxmax()
print(f"The most commonly used ship mode is: {most_common_ship_mode}")
```
