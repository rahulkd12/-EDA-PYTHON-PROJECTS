Q 1: What is the average age of patients admitted?

```python
average_age = df["Age"].mean()
print(f"Average age of patients: {average_age:.2f} years")
```

Q 2: Which medical condition has the highest number of admissions?

```python
top_condition = df["Medical Condition"].value_counts().idxmax()
condition_count = df["Medical Condition"].value_counts().max()
print(f"Most common medical condition: {top_condition} (Count: {condition_count})")
```

Q 3: What is the average billing amount for patients with different insurance providers?

```python
average_billing = df.groupby("Insurance Provider")["Billing Amount"].mean()
print("Average billing amount by insurance provider:")
print(average_billing)
```

Q 4: How many patients stayed in the hospital for more than 7 days?

```python
# Calculating length of stay
df["Discharge Date"] = pd.to_datetime(df["Discharge Date"])
df["Date of Admission"] = pd.to_datetime(df["Date of Admission"])
df["Length_of_Stay"] = (df["Discharge Date"] - df["Date of Admission"]).dt.days

long_stay_count = df[df["Length_of_Stay"] > 7].shape[0]
print(f"Number of patients with hospital stays longer than 7 days: {long_stay_count}")
```

Q 5: Which doctor has treated the highest number of patients?

```python
top_doctor = df["Doctor"].value_counts().idxmax()
patient_count = df["Doctor"].value_counts().max()
print(f"Doctor with the highest number of patients: {top_doctor} (Count: {patient_count})")
```

```python
# Example 2: Pie chart for 'Gender' distribution
gender_count = df['Gender'].value_counts()
plt.figure(figsize=(8, 8))
gender_count.plot.pie(autopct='%0.2f%%', startangle=90, cmap='Set3')
plt.title('Gender Distribution of Patients')
plt.ylabel('')
plt.show()
```
