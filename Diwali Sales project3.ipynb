Q 1: What is the total amount spent by all customers?

```python
total_amount_spent = df["Amount"].sum()
print(f"Total amount spent by all customers: ${total_amount_spent:.2f}")
```

Q 2: Which product category has the highest number of orders?

```python
top_product_category = df.groupby("Product_Category")["Orders"].sum().idxmax()
total_orders_top_category = df.groupby("Product_Category")["Orders"].sum().max()
print(f"Top product category by orders: {top_product_category} (Orders: {total_orders_top_category})")
```

Q 3: What is the average amount spent by different age groups?



```python
average_amount_by_age_group = df.groupby("Age Group")["Amount"].mean()
print("Average amount spent by age group:")
print(average_amount_by_age_group)
```

Q 4: How many unique customers are there in the dataset?

```python
unique_customers = df["User_ID"].nunique()
print(f"Number of unique customers: {unique_customers}")
```

Q 5: Which state generated the highest total sales?

```python
top_state = df.groupby("State")["Amount"].sum().idxmax()
state_sales = df.groupby("State")["Amount"].sum().max()
print(f"State with highest total sales: {top_state} (${state_sales:.2f})")
```
